# Bài toán N quân Hậu (N-Queens)

Đặt N quân hậu trên bàn cờ NxN sao cho không có hai quân hậu nào tấn công nhau (cùng hàng, cột, hoặc đường chéo).  
Thuật toán **backtracking**: đặt hậu lần lượt theo từng hàng, tại mỗi hàng thử từng cột và kiểm tra ràng buộc. Nếu gặp ngõ cụt thì quay lui thử cột khác.

```python
class NQueensSolver:
    def __init__(self, n):
        self.n = n
        self.positions = [-1] * n  # positions[row] = col
        self.solution_count = 0

    def solve_and_print(self):
        """Tìm và in tất cả các cách xếp hậu."""
        self.solution_count = 0
        self._backtrack(0)
        print(f"\n>> Tổng cộng {self.solution_count} cách xếp cho bàn cờ {self.n}x{self.n}.")

    def _is_safe(self, row, col):
        """Kiểm tra (row, col) có bị hậu ở các hàng trước tấn công không."""
        for prev_row in range(row):
            prev_col = self.positions[prev_row]
            # Trùng cột hoặc cùng đường chéo
            if prev_col == col or abs(row - prev_row) == abs(col - prev_col):
                return False
        return True

    def _backtrack(self, row):
        """Đệ quy đặt hậu từng hàng bằng backtracking."""
        if row == self.n:
            self.solution_count += 1
            self._print_solution()
            return

        for col in range(self.n):
            if self._is_safe(row, col):
                self.positions[row] = col
                self._backtrack(row + 1)

    def _print_solution(self):
        """In bàn cờ của một giải pháp."""
        print(f"Giải pháp #{self.solution_count}:")
        for row in range(self.n):
            line = ""
            for col in range(self.n):
                line += "Q " if self.positions[row] == col else ". "
            print(line.strip())
        print("-" * (2 * self.n - 1))
```

```python
print("--- BÀN CỜ 4x4 ---")
NQueensSolver(4).solve_and_print()

print("\n" + "=" * 40 + "\n")

print("--- BÀN CỜ 8x8 ---")
NQueensSolver(8).solve_and_print()
```
