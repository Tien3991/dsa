```python
# ===== Bài 1: Cặp điểm gần nhất trong 2D (O(n log n)) =====

def closest_pair(points):
    if len(points) < 2:
        return float("inf"), (None, None)

    pts = sorted(points, key=lambda p: (p[0], p[1]))

    def dist2(a, b):
        dx = a[0] - b[0]
        dy = a[1] - b[1]
        return dx*dx + dy*dy

    def merge_by_y(left, right):
        out = []
        i = j = 0
        while i < len(left) and j < len(right):
            if left[i][1] <= right[j][1]:
                out.append(left[i]); i += 1
            else:
                out.append(right[j]); j += 1
        if i < len(left):  out.extend(left[i:])
        if j < len(right): out.extend(right[j:])
        return out

    def solve(arr):
        n = len(arr)
        if n <= 3:
            best = float("inf")
            pair = (None, None)
            arr_y = sorted(arr, key=lambda p: p[1])
            for i in range(n):
                for j in range(i+1, n):
                    d2 = dist2(arr[i], arr[j])
                    if d2 < best:
                        best = d2; pair = (arr[i], arr[j])
            return best, pair, arr_y

        mid = n // 2
        midx = arr[mid][0]
        dl, pair_l, yl = solve(arr[:mid])
        dr, pair_r, yr = solve(arr[mid:])
        d = dl; pair = pair_l
        if dr < d:
            d = dr; pair = pair_r

        y_sorted = merge_by_y(yl, yr)
        strip = [p for p in y_sorted if (p[0] - midx)**2 < d]

        # Mỗi điểm so sánh tối đa 7 điểm tiếp theo trong strip (đã sắp theo y)
        m = len(strip)
        for i in range(m):
            # kiểm tra các điểm tiếp theo
            t = 1
            j = i + 1
            while j < m and t <= 7:
                d2 = dist2(strip[i], strip[j])
                if d2 < d:
                    d = d2; pair = (strip[i], strip[j])
                j += 1; t += 1

        return d, pair, y_sorted

    best2, pair, _ = solve(pts)
    return best2, pair

def closest_pair_distance(points):
    """Khoảng cách thật (sqrt) nếu cần."""
    d2, pair = closest_pair(points)
    return (d2 ** 0.5, pair)
```

```python
# ===== Bài 2: Nhân số nguyên lớn m chữ số (không dùng int lớn sẵn có) =====

def multiply_bigint(a: str, b: str) -> str:
    """
    Nhân hai số nguyên lớn biểu diễn bằng chuỗi thập phân.
    Hỗ trợ dấu. Thuật toán nhân truyền thống O(n*m).
    """
    # loại bỏ khoảng trắng
    a = a.strip()
    b = b.strip()
    if not a or not b:
        return "0"

    # dấu
    sign = 1
    if a[0] == '-':
        sign *= -1
        a = a[1:]
    if b[0] == '-':
        sign *= -1
        b = b[1:]

    # xóa số 0 đầu
    a = a.lstrip('0') or "0"
    b = b.lstrip('0') or "0"
    if a == "0" or b == "0":
        return "0"

    n, m = len(a), len(b)
    res = [0] * (n + m)

    # nhân: từ phải sang trái
    ra = [ord(c) - 48 for c in a[::-1]]  # chữ số đảo
    rb = [ord(c) - 48 for c in b[::-1]]

    for i in range(n):
        carry = 0
        ai = ra[i]
        for j in range(m):
            total = res[i + j] + ai * rb[j] + carry
            res[i + j] = total % 10
            carry = total // 10
        if carry:
            res[i + m] += carry

    # loại bỏ 0 thừa ở đầu mảng kết quả (cuối khi đảo)
    while len(res) > 1 and res[-1] == 0:
        res.pop()

    out = ''.join(chr(d + 48) for d in res[::-1])
    if sign < 0:
        out = "-" + out
    return out
```

```python
if __name__ == "__main__":
    # Bài 1
    pts = [(0,0), (5,5), (3,4), (3,3), (1,1), (2,2)]
    d2, pair = closest_pair(pts)
    print("Bài 1: d^2 =", d2, "cặp =", pair)
    d, pair = closest_pair_distance(pts)
    print("Bài 1: d   =", d, "cặp =", pair)

    # Bài 2
    x = "12345678901234567890"
    y = "98765432109876543210"
    print("Bài 2:", multiply_bigint(x, y))
```
