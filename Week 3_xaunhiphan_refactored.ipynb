# Sinh xâu nhị phân không có 2 bit 1 liên tiếp

Cho số nguyên dương `n` (n < 10), in ra tất cả các xâu nhị phân độ dài `n` sao cho không có hai bit `1` nào đứng liền kề nhau.  
Sử dụng thuật toán **backtracking** để duyệt và cắt tỉa các nhánh không hợp lệ.

```python
def generate_valid_binary_strings(n):
    """Sinh và in tất cả xâu nhị phân hợp lệ độ dài n."""
    if n <= 0:
        print("Độ dài n phải là số nguyên dương.")
        return

    print(f"--- Các xâu nhị phân hợp lệ độ dài {n} ---")
    _backtrack(n, "")


def _backtrack(n, current_string):
    """Đệ quy sinh xâu nhị phân bằng backtracking."""
    # Đủ độ dài -> in kết quả
    if len(current_string) == n:
        print(current_string)
        return

    # Luôn có thể thêm '0'
    _backtrack(n, current_string + '0')

    # Chỉ thêm '1' nếu ký tự cuối không phải '1'
    if len(current_string) == 0 or current_string[-1] == '0':
        _backtrack(n, current_string + '1')
```

```python
generate_valid_binary_strings(4)
print()
generate_valid_binary_strings(8)
```
